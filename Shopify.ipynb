###Question a
Orders from Store_id 42 with order amount of 70400 were all done by the same customer, and each order quantity is fairly large of 2000 pairs. The orders all happens at the same time of the day, thus, it is very likely to be a vendor purchase.

Average order amount from Store 78 has value of 49213.04, and it has no order with order amount less than 25725. 

If we want to calculate the AOV, the orders with order amount of 704000 from Store 42 and the orders from Store 78 should be removed for a better representation of central tendency. 

```python
##Drop data from Store_id 78, and orders amount of $70400 orders from Store 42
df.drop(df[(df.order_amount == 704000) & (df.shop_id==42)].index,inplace=True)
df.drop(df[df.shop_id==78].index,inplace=True)
```

```python
##Part b
SELECT Orders.EmployeeID, Employees.LastName,COUNT(*) FROM Orders
INNER JOIN Employees ON Employees.EmployeeID=Orders.EmployeeID
GROUP BY Orders.EmployeeID 
ORDER BY COUNT(*) DESC
LIMIT 1
```

```python
##Part c
with cte1 as (SELECT Orders.OrderID, Orders.CustomerID, Customers.Country,OrderDetails.ProductID,  OrderDetails.Quantity  FROM 
Orders INNER JOIN Customers ON Orders.CustomerID=Customers.CustomerID
INNER JOIN OrderDetails ON Orders.OrderID=OrderDetails.OrderID
WHERE COUNTRY='Germany'),
cte2 as (SELECT ProductID, SUM(Quantity) as Quantity FROM cte1 GROUP BY ProductID ORDER BY Quantity DESC LIMIT 1)
SELECT Products.ProductName, cte2.Quantity FROM cte2 JOIN Products ON cte2.ProductID=Products.ProductID
```

Germany customers orders Boston Crab Meat the most of with quantity of 160
